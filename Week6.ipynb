```python
import numpy as np
import pandas as pd
import cv2
import glob
from collections import Counter
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.models import load_model
```

```python
link = 'dogs-vs-cats/train'
link = glob.glob(link + '/*')
link
```

```python
# láº¥y label tá»« ÄÆ°á»ng dáº«n
def get_label(link):
    label = []
    for i in link:
        if 'dog' in i.split('\\')[-1]:
            label.append(1)
        else:
            label.append(0)
    return label

# Äá»c áº£nh tá»« ÄÆ°á»ng dáº«n vÃ  tiá»n xá»­ lÃ½ áº£nh
def get_train(link):
    img = []
    for i in link:
        image = cv2.imread(i)
        image = cv2.resize(image, (150, 150))
        image = image / 255.0
        img.append(image)
    return img
```

```python
X_train = np.array(get_train(link))
y_train = np.array(get_label(link))
```

```python
# xÃ¢y dá»±ng mÃ´ hÃ¬nh
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# training
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
model.save('model.h5')
```

```python
# load model
model = load_model('model.h5')

# dá»± ÄoÃ¡n
link = 'dogs-vs-cats/test'
link = glob.glob(link + '/*')
X_test = np.array(get_train(link))
y_pred = model.predict(X_test)
y_pred = [1 if i > 0.5 else 0 for i in y_pred]
```

```python
# load ground truth
link = 'dogs-vs-cats/sampleSubmission.csv'
df = pd.read_csv(link)

# so sÃ¡nh káº¿t quáº£ dá»± ÄoÃ¡n vÃ  ground truth
y_true = df['label']
print('Accuracy:', np.mean(y_true == y_pred))
print('Confusion matrix:', Counter(y_true == y_pred))
```
